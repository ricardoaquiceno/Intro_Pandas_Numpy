##Data Series

```python
#pd.Series(List,index=idex_list) Allows me to create a series with the mentioned indexes. If I do not give Any indexes
# the series automatically assigns indexes as an ndarray
psg_players= pd.Series(['Navas','Mbappe','Neymar','Messi'],index=[1,7,10,30])
series_no_index=pd.Series(['This','is','a','test'])
print(psg_players)
print(series_no_index)
```

```python
#We can also create series from dictionanies as pd.Series(dictionary) as the dictionaries already have 'Keys' these will work as the indexes
my_di={1:'Navas',7:'Mbappe',10:'Neymar',30:'Messi'}
series_from_d=pd.Series(my_di)
print(series_from_d)
```

```python
# We can index the series with [index], but be careful with the slicing as for the slicing it takes into account indexes from 0 to n-1 being n the 
#size of the array 
my_di={1:'Navas',7:'Mbappe',10:'Neymar',30:'Messi'}
series_from_d=pd.Series(my_di)
print(series_from_d[1:3]) #as can be seen, this slicing brings hiden or native indexes 1 and 2, which are series indexes 7 and 10. 
print(series_from_d[7:30]) #as you see,this slicing brings nothing
```

# Filtering with conditions

There are other ways also additional to the ones shown below, that s called "boolean indexing"

https://kanoki.org/2020/01/21/pandas-dataframe-filter-with-multiple-conditions/


```python
dictionary={'Col1':[1,2,3,np.nan],
 'Col2':[4,np.nan,6,7],
 'Col3':['a','b','c',None]}
df=pd.DataFrame(dictionary)
df=df.interpolate()
```

```python
# We can use the logic operators to generate conditions, whenever we define daa_fram['Col'] (condition)
# what we are creating is a series with booleans. 
mayor_cuatro=df['Col2']>4
mtres=df['Col1']>2
```

```python
mtres
```

```python
mayor_cuatro
```

```python
df[mayor_cuatro]#Here we are asking only for the ones that are true according to mayor_cuatro
```

```python
# If one wants to nest several conditions in the same filter without creating variables,
# one must put every condition in parenthesis as below
df[(df['Col1']>2) & (df['Col2']>4)]
```

```python
#en Pandas, puedo negar con ~
df[~mtres & mayor_cuatro]
```

# Merge, Join and Concat

###.merge()

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k4'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

```python
#With how='left' we do a left merge which prioritizes the values from the Dataframe
# "izq" bringign them all as can be seen below. On the other hand, uses the 
# values of 'key' from "izq" to compare with the values of 'key' from "der"
# That's why you see that only the values for the keys k0,k1,k2 were brought
# from the DataFrame "der" and the 'key' k4 was left out  
izq.merge(der,on='key',how='left')
```

```python
#With how='right' we do a right merge which prioritizes the values from the Dataframe
# "der" bringign them all as can be seen below. On the other hand, uses the 
# values of 'key' from "der" to compare with the values of 'key' from "izq"
# That's why you see that only the values for the keys k0,k1,k2 were brought
# from the DataFrame "izq" and the 'key' k3 was left out  
izq.merge(der,on='key',how='right')
```

```python
# with how='inner' we do an inner merge wich compares the values of 'key' from
# izq and compares them to  the values of 'key' from der, only the ones who match
# are brought, as you can see below. 
izq.merge(der,on='key',how='inner')
```

```python
# with how='outer' we do an outer merge wich does not compare and just add everything,
# putting Nan to the colums where there are no values available.
izq.merge(der,on='key',how='outer')
```

###.melt(id_vars=,value_vars=)

Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.

This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are "unpivoted" to the row axis, leaving just two non-identifier columns, 'variable' and 'value'.

```python
books=pd.read_csv("/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv")
```

#.apply()

```python
books=pd.read_csv("/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv")
```

```python
def two_times(value):
    return value * 2 
```

```python
# Here, I am adding a new column where I duplicate the user rating
books['Rating_2'] = books['User Rating'].apply(two_times)
```

```python
books.head()
```

```python
#one can also use lambda functions as the parameter of the function, will be the
# value in the column
books['Rating_2'] = books['User Rating'].apply(lambda x : x*3)
```

```python
books.head()
```

```python
#But we can also use the whole dataframe and be creative and use several different syntaxis 
#that take into account other values from the data_frame
# as you can see, only the ratings when 'genre' is fiction were multiplied by 2 and x from the lambda 
# is the whole dataframe.
books['Rating_2'] = books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'],axis=1)
```
